```python
import pandas as pd

# Read the CSV file into a DataFrame
df = pd.read_csv('modified_data1.csv')

# Display the original DataFrame
print("Original DataFrame:")

# Mapping values to numerical values
gender_mapping = {'f': 0, 'm': 1}
df['gender'] = df['gender'].map(gender_mapping)
yes_no_map= {'no': 0, 'yes': 1}
df['austim'] = df['austim'].map(yes_no_map)
df['jundice'] = df['jundice'].map(yes_no_map)
df['age'].fillna(0, inplace=True)
# Display the updated DataFrame
df = df.drop_duplicates()
print("\nUpdated DataFrame:")
print(df)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.preprocessing import StandardScaler

#DataFrame
feature_cols = ['A1_Score', 'A2_Score', 'A3_Score', 'A4_Score', 'A5_Score',
                'A6_Score', 'A7_Score', 'A8_Score', 'A9_Score', 'A10_Score', 'age', 'gender', 'jundice']
target_col = 'class'

X = df[feature_cols]
y = df[target_col]

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)

# Feature Scaling 
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Create and train the Random Forest model
random_forest_classifier = RandomForestClassifier(n_estimators=100, random_state=42)
random_forest_classifier.fit(X_train_scaled, y_train)

# Make predictions on the test set
y_pred = random_forest_classifier.predict(X_test_scaled)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
print("Classification Report:\n", classification_report(y_test, y_pred))
```
