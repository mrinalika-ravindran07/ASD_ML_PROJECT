```python
import pandas as pd

# Read the CSV file into a DataFrame
df = pd.read_csv('modified_data1.csv')

# Display the original DataFrame
print("Original DataFrame:")

# Mapping values to numerical values
gender_mapping = {'f': 0, 'm': 1}
df['gender'] = df['gender'].map(gender_mapping)
yes_no_map= {'no': 0, 'yes': 1}
df['austim'] = df['austim'].map(yes_no_map)
df['jundice'] = df['jundice'].map(yes_no_map)
df['age'].fillna(0, inplace=True)
# Display the updated DataFrame
df = df.drop_duplicates()
print("\nUpdated DataFrame:")
print(df)

```

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import chi2_contingency
from pandas.api.types import CategoricalDtype
import os
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report
```

```python
feature_cols = ['A1_Score','A2_Score','A3_Score','A4_Score','A5_Score',
                'A6_Score','A7_Score','A8_Score','A9_Score','A10_Score','age','gender','jundice','austim'] 
target_col = 'class'
```

```python
X = df[feature_cols]
y = df[target_col]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
```

```python
svm_classifier = SVC(kernel='poly') 
svm_classifier.fit(X_train, y_train)
y_pred = svm_classifier.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
print("Classification Report:\n", classification_report(y_test, y_pred))
```
